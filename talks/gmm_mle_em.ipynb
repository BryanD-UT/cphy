```python
def multivariate_normal_pdf(x, mu, sigma):
    """
    Compute the multivariate normal pdf at x.
    
    Args:
        x (np.ndarray): point at which to evaluate the pdf
        mu (np.ndarray): mean of the distribution
        sigma (np.ndarray): covariance matrix of the distribution
        
    Returns:
        float: pdf value
    """
    x = x - mu
    prefactor = 1.0 / np.sqrt((2 * np.pi) ** len(x) * np.linalg.det(sigma))
    exponent = 0.5 * np.dot(np.dot(x, np.linalg.inv(sigma)), x)
    return prefactor * np.exp(-exponent)


xx, yy = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))



# https://medium.com/@prateek.shubham.94/expectation-maximization-algorithm-7a4d1b65ca55
```
